## Matplotlib

Matplotlib is a powerful plotting library for python that enables the creation of static, animated and interactive visualizations. It is widely used for data visualization in data science and analytics. In this lesson, we will cover the basics of Matplotlib, including creating various types of plots and customizing them>

```python
import matplotlib.pyplot as plt
```

```python
x =[1,2,3,4,5]
y=[1,3,4,5,56]

## create aline plot

plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("Y")
plt.title("Basic Line Plot")
plt.show()
```

```python
x =[1,2,3,4,5]
y=[1,3,4,5,56]

## create customized line plot



plt.plot(x,y,color ="red",linestyle='--',marker="o",linewidth=4,markersize=4)
plt.grid()
plt.show()
```

```python
## Multiple Plot

## Sample Data
x =[1,2,3,4,5]
y1=[1,3,4,5,56]
y2 =[1,2,3,4,5]

plt.figure(figsize=(9,5))
plt.subplot(1,2,1)
plt.plot(x,y1,color='green')

plt.title("Plot 1")

plt.subplot(1,2,2)

plt.plot(y1,x,color="blue")
plt.title("Plot 2")
```

```python
## Bar plot

categories = ["A","B","C","D","E"]
values = [5,6,7,8,5]
```

```python
plt.bar(categories,values,color="black")
```

## histogram
Histogram are used to represent the distribution of a dataset. they divide the data into bins and count the number of data points in each bin.



```python
data =[1,2,4,5,6,7,7,8,9,9,7,7]

## create a histogram

plt.hist(data,bins=6,color="blue",edgecolor="red")
```

```python
##create a scatter plot

x =[1,2,3,4,5,6]
y = [2,3,4,5,6,5]

plt.scatter(x,y,color="blue",marker="x")
```

```python
## piechart

labels =["A","B","C","D"]
sizes = [30,20,40,10]
colors = ["gold","yellowgreen","lightcolor","lightskyblue"]
explode=(0.2,0,0,0) ## move out the first slite

plt.pie(sizes,explode=explode,labels=labels,autopct="%1.1f%%",shadow=True)
```
